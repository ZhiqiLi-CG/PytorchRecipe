1. Test whether is tensor

```python
import torch
x=[12,23,34,45,56,67,78]
print(torch.is_tensor(x))
print(torch.is_storage(x))
```

2. Create tensor

```python
y=torch.randn(1,2,3,4,5)
print(y)
print(torch.is_tensor(y))
print(torch.is_storage(y))
print(torch.numel(y)) # the total number

y=torch.zeros(4,4)
print(y)
print(torch.numel(y))

import numpy as np
x=[12,23,34,45,56,67,78]
x1=np.array(x)
x2=torch.from_numpy(x1)
print(x,x1,x2)

```

3. eye function

4. linspace, logspace and rand

```python
torch.linspace(2,10,steps=25)
torch.logspace(2,10,steps=25)

print(torch.rand(10))
print(torch.rand(4,5))
print(torch.randn(10)) # norm distribution
print(torch.randn(4,5))

print(torch.randperm(10))
print(torch.arange(10,40,2))
print(torch.arange(10,40))
```

5. find the max and min

```python
d=torch.randn(4,5)
print(d)
print(torch.argmin(d,dim=0)) # min in every row
print(torch.argmin(d,dim=1)) # min in every col
```

6. concatenation and chunk and split(for 1D tensor)

```python
x=torch.randn(4,5)
print(x)
print(torch.cat((x,x))) # cat by row
print(torch.cat((x,x),0)) # cat by row
print(torch.cat((x,x),1)) # cat by col

x=torch.randn(4,4)
print(x)
print(torch.chunk(x,2))   # split along row
print(torch.chunk(x,2,0)) # split along row
print(torch.chunk(x,2,1)) # split along col

print(torch.split(torch.tensor([12,23,34,45,56,67]),2))
```

7. gather:  The gather function collects elements from a tensor and places it in
another tensor using an index argument. The index position is determined
by the LongTensor function in PyTorch.
&
select

```python
a=torch.Tensor([[11,12],[23,24]])
b=torch.LongTensor([[0,0],[1,0]])
print(a)
print(b)

print(torch.gather(a,0,b))  # 1: 列保留，选择行
print(torch.gather(a,1,b))  # 1: 行保留，选择列


a=torch.randn(4,4)
print(a)
indices=torch.LongTensor([0,2])
print(indices)
print(torch.index_select(a,0,indices))
print(torch.index_select(a,1,indices))
```

8. nonzero

9.change the shape
- transpose
- unbind(remove a dim)

```python
x=torch.rand(4,5)
print(x)
print(x.t()) # only used for 2D

x=torch.rand(4,5,3)
print(x.transpose(0,2)) #for all dims

x=torch.rand(4,5)
print(x)
print(torch.unbind(x,0)) # remove row
print(torch.unbind(x,1)) # remove col
```

10. Scalar,ceil,floor and clamp

```python
x=torch.rand(4,5)
print(torch.add(x,20))
print(torch.mul(x,2))

torch.manual_seed(1234)
x=torch.randn(5,5)
print(x)
print(torch.ceil(x))
print(torch.floor(x))
print(torch.clamp(x,min=-0.3,max=0.4))
```

11. exp(calculate exp for each element, not the exp of matrix) and get frac portion

```python
x=torch.rand(4,5)
x[0,0]=1
print(torch.exp(x))
print(torch.frac(x))
print(torch.log(x))
print(torch.pow(x,2))
print(torch.sqrt(x))
```
